### HTTP

Significa *HyperText Transfer Protocol* y es un protocolo de aplicacin que transfiere recursos basados en la web, como documentos HTML, entre un *cliente* y un *servidor web*. Los clientes (buscadores web) y los servidoeres se comunican o intercambian informacin utilizando *HTTP requests* y *HTTP responses*.



Este proceso de comunicacin es cclico, y podemos verlo como una serie de "preguntas y respuestas".

Existe otra versin encriptada y ms segura del protocolo HTTP, y es *HyperText Transfer Protocol Secure* (*HTTPS*). Utiliza las tecnologas *Secure Sockets Layer (SSL)* y *Transport Layer Security (TLS)* para comunicar contenido encriptado entre un cliente y un servidor. Este tipo de seguridad permite que los clientes compartan informacin sensible con un servidor de manera segura. 

Actividades como banca en lnea, compras en lnea y pagos electrnicos utilizan puertos HTTPS para mantener esta informacin segura.

- Nota: El URL (Uniform Resource Locators) de una peticin HTTP comienza como `http://`, mientras que una HTTPS comienza como `https://`.

#### HTTP requests

Los buscadores web envan sus peticiones al servidor. Estas son procesadas utilizando varios mtodos, como **GET** y **POST**.

- GET: Es el mtodo ms comn para solicitar informacin. Se considera un mtodo seguro ya que el estado del recurso no se ve alterado. Adems, se usa para dar query strings, por ejemplo:
``https://www.google.com/search?q=world%20cup%20football&source=hp``, es una solicitud de informacin de Google basado en `q`:(world cup football) y `source`: los parmetros enviados con la peticin. La informacin o queries (`q` y `source` en este ejemplo) se muestran en la URL.

- POST: Se utiliza para hacer una peticin segura al servidor. El estado del contenido solicitado puede ser alterado. Los datos enviados al URL solicitados no son visibles en el URL sino en el cuerpo de la respuesta. Se utiliza para enviar informacin al servidor de manera segura, como en inicios de sesin y registros de usuario.

Ms adelante veremos ms de estos mtodos. Existen dos partes principales en la comunicacin HTTP. Con esta idea bsica sobre las peticiones, exploremos las respuestas.

### HTTP responses

El servidor procesa las peticiones, y a veces los encabezados especificados. Cuado las peticiones son recibidas y procesadas, el servidor enva una respuesta al buscador. La mayora de las veces, las respuestas estn en formato HTML, en otras en JavaScript u otro tipo de documentos, en *JavaScript Object Notation* (*JSON*) u otros formatos.

Una respuesta contiene cdigos de status, la siguiente lista contiene algunos cdigos y su significado:

- **200**: OK, peticin exitosa
- **404**: No encontrado, el recurso solicitado no puede ser encontrado.
- **500**: Error interno del servidor.
- **204**: No hay contenido para enviar.
- **401**: Peticin no autorizada.

### HTTP cookies

Las *cookies* son datos enviados por el servidor al buscador. Estos datos son generados y almacenados por sitios en tu computadora. Ayuda a identificar peticiones HTTP del usuario al sitio web y contienen informacin respecto al manejo de inicio de sesiones, preferencias del usuario y su comportamiento.

El servidor se identifica y comunica con el buscador basado en la informacin almacenada en las cookies. Los datos almacenados ah ayudan al sitio web a transferir ciertos valores, como el usuario y contrasea, propociando una interaccin rpida entre la peticin y la respuesta.

### HTTP proxies
 
Un servidor proxy funciona como un intermediario entre un cliente y el servidor web. El buscador web enva peticiones al servidor, pasa por la proxy y la proxy regresa la respuesta del servidor al cliente.

Las proxies se utilizan generalmente para monitoreo de trfico, en la mejora de desempeo y para la seguridad de recursos de internet. 

Cuando no tenemos acceso a sitios disponibles en nuestra ubicacin, (el mensaje de "este contenido no est disponible en tu regin"), podemos acceder cambiando a una proxy distinta.

## HTML

Los sitios web estn hechos de pginas o documentos que contienen texto, imgenes, hojas de estilo y programas, entre otras cosas. Generalmente se construyen con lenguajes de marcado como **HTML**. 

HTML es el lenguaje de marcado estndar para construir una pgina web. Desde principios de 1990, HTML se ha utilizado independientemente y en conjunto con lenguajes basados en servidores como PHP, ASP y JSP. 

HTML define y almacena el contenido de una pgina web. Los datos que buscamos pueden extraerse y ser encontrados dentro de las pginas HTML, dentro de un conjunto de instrucciones predefinidas conocidas como *tags*. Las *tags* son generalmente un marcador para ciertos atributos, por ejemplo `<a>, <b>, <table>, <img>`, etc.

### Elementos y atributos HTML

Los *elementos* HTML (tambin conocidos como nodos de documento) son los bloques de construccin de los documentos web. Los elementos comienzan con una etiqueta inicial `<...>` y una etiqueta final `/<...>` con cierto contenido dentro de ellos.

Un elemento tambin puede contener *atributos*, generalmente definidos como:

`attribute-name = attribute-value`

los cuales dan informacin adicional al elemento.

Por ejemplo:

    <p> Esto es una etiqueta de prrafo </p>

    <h1> Esto es una etiqueta de encabezado!, tambin lo son h2, h3, h4, h5 y h6</h1>

    <a href="https://www.google.com"> Click aqu para ir a Google</a>

    <img src="src\img\doggy.jpg" width="400" height="600" alt="Perrito" />
    
    <br/>

- `<p>` y `<h1>` son elementos HTML que contienen informacin en texto. P de "prrafo", H de "header".

- `<a>` se define como un atributo `href` que contiene el link que ser procesado al hacer click en el texto "Click aqu para ir a Google".

- La etiqueta `<img>` tambin contiene algunos atributos, como `src` y `alt` junto a sus valores. `src` aloja el recurso, mientras que `alt` es para el texto alternativo (generalmente se usa cuando hay una conexin lenta o cuando no es posible cargar la imagen). 

- `<br/>` representa un salto de lnea en HTML y no tiene atributos o contenido de texto. Se utiliza para insertar una nueva lnea en la estructura del documento.

Los elementos de HTML tambin pueden ser anidados en estructuras anidadas, con una jerarqua padre-hijo:

    <div class="article">
        <p id="mainContent" class="content">
            <b>Perrito de Shrek </b>
            <br/>        
            <img src="src\img\Perrito.png" id="pageLogo" alt="Perrito" class="logo"/>
        </p>
        <p>
            <h3> Web Scraping! </h3>
        </p>
    </div>

En el ejemplo anterior, dos elementos hijo `<p>` se encuentran dentro del bloque `<div>`. Ambos elementos hijos tienen ciertos atributos y ciertos elementos en su contenido. Generalemente los documentos HTML tienen la estructura anterior.

### Atributos globales

 Los elementos HTML contienen informacin adicional, como pares key-value. Tambin se conocen como atributos de elementos HTML. Dichos atributos contienen valores y dan identificacin, o contienen informacin adicional que puede ser til en distintos aspectos durante el proceso de scraping, como identificando los elementos web exactos y extraer valores o texto de ellos, as como moverse entre elementos.

### Primeros pasos

Veamos cmo funciona. Primero importamos `requests`.

```python
import requests
```

Vamos a intentar conectarnos a la pgina de documentacin de la biblioteca.

```python
link = "https://requests.readthedocs.io/"
```

Ahora utilizamos el mtodo `get()` para hacer una peticin a la pgina.

```python
type(respuesta)
```

Notemos que `respuesta` es un objeto `Response` de `requests`. Despus de realizar una peticin podemos acceder a distintos atributos de la misma:

```python
codigo_estado = respuesta.status_code

print(codigo_estado)
```

```python
encabezados = respuesta.headers

print(encabezados)
```

```python
codificacion = respuesta.encoding

print(codificacion)
```

Tambin es posible pasar algunos parmetros en la URL con el objetivo de buscar algo en especfico en la pgina.

Supongamos que nos interesa saber cules son los repositorios escritos en Python con mayor cantidad de estrellas en GitHub.

```python
params = {'q': 'python', 'sort': 'stars'}

best_repos = requests.get('https://api.github.com/search/repositories', params=params)
```

Adicionalmente, para mejor lectura podemos generar un archivo con la informacin obtenida:

```python
f = open('mejores_repositorios.txt', "wb")
f.write(best_repos.content)
f.close()
```
